```python
import os
import chromadb
from openai import OpenAI
from dotenv import load_dotenv
from chromadb.utils.embedding_functions import OpenAIEmbeddingFunction
```

```python
#getting the api key
load_dotenv()
Api_key = os.getenv("OPENAI_API_KEY")
openai = OpenAI(api_key=Api_key)
#chroma client
client = chromadb.PersistentClient(path="./chromdb")
```

```python
resumes = [
    {
        "id": "1",
        "name": "Alice Johnson",
        "skills": "Python, Django, PostgreSQL",
        "experience": "3 years building web applications.",
        "education": "BSc Computer Science"
    },
    {
        "id": "2",
        "name": "Bob Smith",
        "skills": "Java, Spring Boot, MySQL",
        "experience": "5 years developing enterprise software.",
        "education": "BSc Software Engineering"
    },
    {
        "id": "3",
        "name": "Carol Williams",
        "skills": "Python, AWS, Docker, Kubernetes",
        "experience": "4 years building cloud applications.",
        "education": "MSc Computer Science"
    },
    {
        "id": "4",
        "name": "David Brown",
        "skills": "React, JavaScript, Node.js",
        "experience": "2 years as a full-stack developer.",
        "education": "BSc Information Technology"
    },
    {
        "id": "5",
        "name": "Emma Davis",
        "skills": "Python, Machine Learning, TensorFlow",
        "experience": "3 years developing AI solutions.",
        "education": "MSc Artificial Intelligence"
    }
]
```

```python
# OpenAI embedding function
embedding_function = OpenAIEmbeddingFunction(
    api_key=Api_key ,
    model_name="text-embedding-3-small"
)
```

```python
#deleting old collection to add a new one 
try:
    client.delete_collection("movies")
    print("Old collection deleted.")
except:
    print("Collection doesn't exist yet.")
```

```python
collection =client.get_or_create_collection(
    name="resume",
    embedding_function=embedding_function
)
```

```python
#loading data
metadatas = []
documents = []
ids = []

for resume in resumes:
    documents.append(
        f"""
        Name: {resume['name']}
        Answer: {resume['education']}
        Experience: {resume['experience']}
        """
    )

    ids.append(resume["id"])

    metadatas.append({
        "skills": resume["skills"],
    })
```

```python
collection.add(
    ids=ids,
    documents=documents,
    metadatas=metadatas
)
```

```python
result = collection.query(
    query_texts=['David Brown'],
    n_results=1
)
```
